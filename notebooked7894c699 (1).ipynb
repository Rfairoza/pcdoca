```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
lego_a = cv2.imread('../input/fotoini/photo_2022-07-04_16-35-19.jpg',1)
lego_a = cv2.cvtColor(lego_a, cv2.COLOR_BGR2RGB)
lego_b = cv2.imread('../input/fotoitu/photo_2022-07-04_16-35-09.jpg',1)
lego_b = cv2.cvtColor(lego_b, cv2.COLOR_BGR2RGB)
lego_blending = cv2.addWeighted(lego_a, 0.7, lego_b, 0.5,0)
plt.figure(figsize=(20, 20))
plt.subplot(1,3,1)
plt.imshow(lego_a)
plt.subplot(1,3,2)
plt.imshow(lego_b)
plt.subplot(1,3,3)
plt.imshow(lego_blending)
plt.show()
```

```python
from PIL import Image, ImageEnhance
#read the image
im = Image.open("../input/fotoitu/photo_2022-07-04_16-35-09.jpg")
#image brightness enhancer
enhancer = ImageEnhance.Brightness(im)
factor = 1 #gives original image
im_output = enhancer.enhance(factor)
im_output.save('asli.png')
factor = 0.5 #darkens the image
im_output = enhancer.enhance(factor)
im_output.save('gelap.png')
factor = 2.5 #brightens the image
im_output = enhancer.enhance(factor)
im_output.save('cerah.png')
plt.show()
```

```python
img = cv2.imread('../input/fotoitu/photo_2022-07-04_16-35-09.jpg',0)
gamma = 2.5
img_power = img**gamma
plt.figure(figsize=(20, 20))
plt.subplot(1,2,1)
plt.imshow(img,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_power,cmap="gray")
plt.show()
```

```python
img = cv2.imread('../input/fotoitu/photo_2022-07-04_16-35-09.jpg',1)
kernel = np.ones((3, 3), np.float32) / 9
img_out = cv2.filter2D(img, -1, kernel)
plt.figure(figsize=(20, 20))
plt.subplot(1,2,2)
plt.imshow(img,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_out,cmap="gray")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img = cv2.imread('../input/fotoitu/photo_2022-07-04_16-35-09.jpg',0)
img_negative = 255 - img
plt.figure(figsize=(15, 15))
plt.subplot(1,2,1)
plt.imshow(img,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_negative,cmap="gray")
plt.show()
```

```python
img = cv2.imread('../input/fotoitu/photo_2022-07-04_16-35-09.jpg',0)
c = 255 / np.log10(1 + np.max(img))
img_log = c * (np.log10(img + 1))
img_log = np.array(img_log, dtype = np.uint8)
plt.figure(figsize=(20, 20))
plt.subplot(1,2,1)
plt.imshow(img,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_log,cmap="gray")
plt.show()
```

```python
img = cv2.imread('../input/fotoitu/photo_2022-07-04_16-35-09.jpg',1)
kernel = np.array([
                   [0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]
])
img_out = cv2.filter2D(img, -1, kernel)
plt.figure(figsize=(20, 20))
plt.subplot(1,2,1)
plt.imshow(img,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_out,cmap="gray")
plt.show()
```
